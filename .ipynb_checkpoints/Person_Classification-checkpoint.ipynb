```python
Face = r"D:\PROJECTS\IMAGE_CLASIFICATION_WEB\model\opencv\haarcascades\haarcascade_frontalface_default.xml"
Eye = r"D:\PROJECTS\IMAGE_CLASIFICATION_WEB\model\opencv\haarcascades\haarcascade_eye.xml"
```

```python
Face_cascade_mark = cv2.CascadeClassifier(Face)
Eye_cascade_mark = cv2.CascadeClassifier(Eye)
```

```python
cv2.destroyAllWindows()
for (x,y,w,h) in Faces:
    Face_img = cv2.rectangle(image,(x,y),(x+w,y+h),(255,0,0),3)
    Roi_grey = gray[y:y+h,x:x+w]
    Roi_Color = Face_img[y:y+h,x:x+w]
    Eyes = Eye_cascade_mark.detectMultiScale(Roi_grey)
    for (a,b,c,d) in Eyes:
        cv2.rectangle(Roi_Color,(a,b),(a+c,b+d),(0,255,0),3)

plt.figure()
plt.imshow(Face_img)
plt.show()
```

```python
plt.imshow(Roi_Color)
```

```python
def Preprocessing_Crop(path):
    img = cv2.imread(path)
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces = Face_cascade_mark.detectMultiScale(gray,1.3,5)
    
    for (x,y,w,h) in faces:
        roi_grey = gray[y:y+h,x:x+w]
        roi_Color = img[y:y+h,x:x+w]
        Eyes = Eye_cascade_mark.detectMultiScale(roi_grey)
        if(len(Eyes)==2):
            return roi_Color
```

```python
check = Preprocessing_Crop(r"D:\PROJECTS\DATASET\py\DataScience\CelebrityFaceRecognition\model\test_images\sharapova1.jpg")
plt.imshow(check)
plt.show()
```

```python
dataset_path = r"D:\PROJECTS\IMAGE_CLASIFICATION_WEB\model\dataset"
crop_path = r"D:\PROJECTS\IMAGE_CLASIFICATION_WEB\model\dataset\cropped"
```

```python
import os
image_dirs = []
for entry in os.scandir(dataset_path):
    if entry.is_dir():
        image_dirs.append(entry.path)

image_dirs
```

```python
import shutil
if os.path.exists(crop_path):
    shutil.rmtree(crop_path)
os.makedirs(crop_path)
```

```python
cropped_temp = []
name_filenames_dict = {}#contains all the file ka file path 

for dirs in image_dirs:
    count = 1
    name = dirs.split('\\')[-1]
    print(name)
    
    name_filenames_dict[name]=[]
    
    for images in os.scandir(dirs):
        crp_img = Preprocessing_Crop(images.path)
        if crp_img is not None:
            cropped_folder = crop_path +'/'+ name
            if not os.path.exists(cropped_folder):
                os.mkdir(cropped_folder)
                cropped_temp.append(cropped_folder)
            
            file_name = name + str(count) +'.png'
            path_name = cropped_folder + "/"+file_name
            
            count+=1
            cv2.imwrite(path_name,crp_img)
            name_filenames_dict[name].append(path_name)
```
